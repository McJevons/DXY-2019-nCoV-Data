```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data=pd.read_csv('csv/DXYArea.csv')

data_copy=data.copy()

data=data.drop(index=data[data['provinceName'] != '湖北省'].index,
          columns=data.columns[[0, 2, 3, 4, 5, 7]])

data['updateTime']=data['updateTime'].map(lambda x:x[:10])
```

```python
def get_data_by_city(data,cityName):
    data=data[data['cityName'].isin([cityName])]
    data=data.groupby('updateTime').max()
    return data
```

```python
data_hs=get_data_by_city(data,'黄石')
```

```python
data_wh=get_data_by_city(data,'武汉')
```

```python
pd.merge(data_hs,data_wh,on='updateTime',how='outer')
```

```python
pd.concat([data_hs,data_wh],keys=['ab','cd'],sort=True)
```
