```python
%matplotlib inline
import numpy as np
import pandas as pd
import json
import requests
import time
from matplotlib import pyplot as plt
```

```python
r=requests.get('https://lab.isaaclin.cn/nCoV/api/area')
```

```python
j=json.loads(r.text)
```

```python
china=[x for x in j['results'] if x['country']=='中国']
```

```python
city=[]
for province in china:
    for cities in province['cities']:
        cities['province']=province['provinceShortName']
        cities['updateTime']=province['updateTime']
        city.append(cities)
```

```python
df=pd.DataFrame(city)
```

```python
df.drop(['locationId'],inplace=True,axis=1)
```

```python
df['updateTime']=df['updateTime'].map(lambda x:time.strftime('%Y-%m-%d',time.localtime(x/1000)))
```

```python
df
```

```python
z=df.groupby('province').sum().sort_values('confirmedCount',ascending=False)
```

```python
z
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
```

```python
plt.figure(figsize=(10,4))
plt.bar(x=z.index[1:],height=z['confirmedCount'][1:])
```
