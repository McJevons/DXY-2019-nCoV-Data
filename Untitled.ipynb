```python
%matplotlib notebook
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 中文转换  
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题

# 创建画板
plt.Figure((8,4))

# 绘制折线图，label是小图标的注解
plt.plot([1,2,3,4],[4,5,3,15],label='第一条折线')
plt.plot([2,3,4],[10,14,12],label = '第二条折线')

# 设置轴标签
plt.xlabel('X轴')
plt.ylabel('Y轴')

# 为图表添加标题
plt.title('简单的折线图')

#确定小图标的位置
plt.legend(loc = 'upper right')   
```

```python
data = pd.read_csv('DXYArea.csv')

data['updateTime'] = data['updateTime'].map(lambda x: x[:10])

data_wuhan = data[data['cityName'] == '武汉']

data_wuhan = data_wuhan.groupby('updateTime').max()

data_wuhan = data_wuhan[[
    'cityName', 'city_confirmedCount', 'city_curedCount', 'city_deadCount'
]]

data_wuhan
```

```python
import seaborn as sns
```

```python
sns.barplot(data=data_wuhan,x=data_wuhan.index,y='city_confirmedCount')
```
