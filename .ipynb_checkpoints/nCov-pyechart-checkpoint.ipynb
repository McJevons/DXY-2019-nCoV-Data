```python
import numpy as np
import pandas as pd
from pyecharts.charts import *
from pyecharts import options as opts
```

```python
from nCovData import nCovData
```

```python
data = nCovData('csv/DXYArea.csv')
```

```python
cities = ['黄石', '鄂州', '黄冈', '孝感']
item = 'city_increase_confirmed'
```

```python
df_cities = data.get_df_by_cities(*cities)
```

```python
def make_chart(chart_type, data, item, cities):
    chart = eval(chart_type)()
    chart.add_xaxis(data.index.to_list())
    for city in cities:
        chart.add_yaxis(city, data.loc[:, (item, city)].to_list())
    return chart
```

```python
bar = make_chart('Bar', df_cities, item, cities)
```

```python
line = make_chart('Line', df_cities, item, cities)
```

```python
chart = bar.overlap(line)
```

```python
chart.render_notebook()
```

```python
df_provinces = data.get_full_df().iloc[:, [0, 2, 4, 5]].groupby(
    'provinceName').max()
```

```python
china_map = Map()
```

```python
def short_privonceName(provinceName):
    del_words = ['市', '省', '自治区', '回族', '壮族', '维吾尔']
    for word in del_words:
        provinceName = provinceName.replace(word, '')
    return provinceName
```

```python
x = df_provinces['province_confirmedCount'].to_dict()
```

```python
china_map.add(
    '确诊人数',
    [list(z) for z in zip(map(short_privonceName, x.keys()), x.values())],
    'china',
    is_map_symbol_show=False)
```

```python
china_map.set_global_opts(
    title_opts=opts.TitleOpts(title="Map-VisualMap（分段型）"),
    visualmap_opts=opts.VisualMapOpts(max_=1500, is_piecewise=True,    pieces=[
                 {"min": 0, "max": 50, "label": "0-50", "color": "red"},
                 {"min": 51, "max": 100, "label": "50-100", "color": "blue"},
                 {"min": 101, "max": 500, "label": "100-500", "color": "yellow"},
                 {"min": 501, "max": 1000, "label": "500-1000", "color": "green"},
                 {"min": 1001, "max": 2000, "label": "1000-2000", "color": "blue"},
                 {"min": 2001, "max": 100000, "label": ">2000", "color": "cyan"}
             ])
)
```

```python
china_map.render_notebook()
```
